```python
df['After 5 Days']=df['Date']+pd.Timedelta(days=5)
```

```python
df['Before_weeK']=df['Date']+pd.Timedelta(weeks=-1)
```

```python
df['After_month']=df['Date']+pd.DateOffset(months=1)
```

```python
df['Before_year']=df['Date']+pd.DateOffset(years=-1)
```

```python
df['time_later']=df['time']+pd.Timedelta(hours=2)
```

```python
df['days_between']=df['After_week']-df['Before_weeK']
```
